```python
import pandas as pd   
import requests       
from bs4 import BeautifulSoup 
import numpy as np 
import re
```

```python
movieTitles = []
year = []
time = []
genre=[]
rating = []
metascore = []
votes = []
gross = []
description = []
Director = []
Stars = []
```

```python
def remove(string):
    return string.replace(" ", "")
```

```python
ct=1
for i in range (4):

    url = 'https://www.imdb.com/search/title/?title_type=feature&num_votes=50000,&sort=user_rating,desc&start='+str(ct)+'&ref_=adv_nxt'
    ct=ct+50
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    movie_data = soup.findAll('div', attrs= {'class': 'lister-item mode-advanced'})

    for store in movie_data:
        name = store.h3.a.text
        movieTitles.append(name)
    
        year_of_release = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '')
        year.append(year_of_release)
    
        try:
            runtime = store.p.find('span', class_ = 'runtime').text.replace(' min', '')
        except Exception as e:
            try:
                runtime = soup.find('span', {'class': 'runtime'}).text
            except Exception as e:
                runtime = ""


        time.append(runtime)
        try:
            cat = store.p.find('span', class_ = 'genre').text
        except Exception as e:
            try:
                cat = soup.find('span', {'class': 'genre'}).text
            except Exception as e:
                cat = ""

        cat=remove(cat)
        cat.strip()
        genre.append(cat.replace("\n",""))
    
        rate = store.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n', '')
        rating.append(rate)
    
        meta  = store.find('span', class_ = 'metascore').text.replace(' ', '') if store.find('span', class_ = 'metascore') else '^^^^^^'
        metascore.append(meta)

        value = store.find_all('span', attrs = {'name': 'nv'})
    
        vote = value[0].text
        votes.append(vote)
    
        grosses = value[1].text if len(value) >1 else '*****'
        gross.append(grosses)
    
        describe = store.find_all('p', class_ = 'text-muted')
        description_ = describe[1].text.replace('\n', '') if len(describe) >1 else '*****'
        description.append(description_)
    
        cast = store.find("p", class_ = '')
        cast = cast.text.replace('\n', '').split('|')
        cast = [x.strip() for x in cast]
        cast = [cast[i].replace(j, "") for i,j in enumerate(["Director:", "Stars:"])]
        Director.append(cast[0])
        Stars.append([x.strip() for x in cast[1].split(",")])
```

```python
movie_DF1 = pd.DataFrame({'Name of movie': movieTitles, 'Year of release': year, 'Watchtime': time,'Genre' : genre,'Movie Rating': rating, 'Metascore': metascore, 'Votes': votes, 'Gross collection': gross, 'Description': description, "Director": Director, 'Star': Stars})
```

```python
def remove_prefix(df, column_name, prefix):
    df[column_name] = df[column_name].str.replace(prefix, '')
    return df
movie_DF1= remove_prefix(movie_DF1, 'Director', 'Directors:')
```

```python
def remove_pipe_regex(df, column_name):
    df[column_name] = df[column_name].apply(lambda x: re.sub('I', '', x))
    return df
movie_DF1 = remove_pipe_regex(movie_DF1, 'Year of release')
```

```python
def comma_separated_to_list(df, column_name):
    df[column_name] = df[column_name].str.split(',').apply(lambda x: [y.strip() for y in x])
    return df
movie_DF1 = comma_separated_to_list(movie_DF1, 'Director')
```

```python
def split_values_to_list(df, column_name):
    df[column_name] = df[column_name].apply(lambda x: x.split(','))
    return df
movie_DF1 = split_values_to_list(movie_DF1, 'Genre')
```

```python
def get_value_counts(df, column_name):
    value_counts = df[column_name].value_counts()
    return value_counts
value_counts = get_value_counts(movie_DF1,'Year of release')
print(value_counts)
```
